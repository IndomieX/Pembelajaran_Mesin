```python
#IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('botak.csv')
```

```python
df
```

```python
#Mengecek Missing Value
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["umur"], bins=30, kde=True, color="black", edgecolor="white")
plt.xlabel("Usia")
plt.ylabel("Jumlah Responden")
plt.title("Distribusi Usia Responden (Rafi Daniswara P.W)", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
gender_counts = df["jenis_kelamin"].dropna().value_counts()

plt.figure(figsize=(6, 6))
colors = ["gold", "blue"]
plt.pie(gender_counts, labels=gender_counts.index, autopct="%.2f%%", colors=colors, startangle=90)
plt.title("Proporsi Jenis Kelamin Responden (Rafi Daniswara P.W)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["gaji"], y=df["botak_prob"], alpha=0.6, color="blue", s=50)
plt.xlabel("Pendapatan")
plt.ylabel("Peluang Kebotakan")
plt.title("Korelasi antara Pendapatan dan Peluang Kebotakan (Rafi Daniswara P.W)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
